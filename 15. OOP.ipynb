```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    def display(self):
        print(self.brand)
        print(self.model)
c = Car("Tata", "1221")
c.display()
        
        
```

## Encapsulation

```python
class Bank:
    def __init__(self,balance):
        self.balance = balance 

    def getBalance(self):
        return self.balance

    def setBalance(self, deposit):
        self.balance += deposit
b = Bank(30000)
b.setBalance(10000)
print(b.getBalance())
```

## inheritance

```python
class Animal:
    def sound(self):
        print("animals sound")
class Dog(Animal):
    def speak(self):
        print("bark")
d = Dog()
d.sound()
d.speak()
```

## polymorphism

```python
class A:
    def display(self):
        print("A")
class B(A):
    def display(self):
        print("B")
a1 = A()
b1 = B()
a1.display()
b1.display()
        
```
