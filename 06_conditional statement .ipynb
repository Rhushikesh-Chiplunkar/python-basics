```python
# elif
if a >= b:
    print("a")
elif b >= c:
    print("b")
else: 
    print("c")
```

```python
# if else for data science
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Score': [85, 40, 75]}
df = pd.DataFrame(data)
df['Result'] = df['Score'].apply(lambda x: 'Pass' if x >= 50 else 'Fail')
print(df)
```
