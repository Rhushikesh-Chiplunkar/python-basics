```python
def http_status(code):
    match code:
        case 200:
            return "ok"
        case 400:
            return "bad requiest"
        case 404:
            return "not found"
        case 500:
            return "internal server error"
        case _:
            return "unknown error"
    
print(http_status(404))
print(http_status(200))
```
